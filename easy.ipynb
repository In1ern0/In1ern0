Вспомогательные функции к задачам

```python
def is_palindrom(n):
    """
    Проверяет является ли число палиндромом
    """
    if str(n) == str(n)[::-1]:
        return True
    return False

def is_prime(n):
    """
    Проверяет является ли число простым
    """
    if n == 2: return True
    elif n == 0 or n == 1 or n % 2 == 0: return False
    for i in range(3, round(n ** 0.5) + 1, 2):
        if n % i == 0:
            return False
    return True

def is_all_prime(n):
    """
    Проверяет являются ли все циклические перестановки простыми числами
    """
    arr = []
    arr.append(n)
    n = str(n)
    if len(n) > 1:
        for i in range(len(n) - 1):
            n = n[1:] + n[0]
            arr.append(int(n))
    for i in arr:
        if is_prime(i) == False:
            return False
    return True
```

Задача 1

*   Найдите все палиндромы $a < 10^5$, такие что $a^2$ — также
палиндром.
*   Найдите все простые числа $p < 10^6$, все циклические пе-
рестановки цифр которых также являются простыми.

```python
def palindromic_squares_and_circular_primes() -> tuple[List[int], List[int]]:
    """
    Возвращает:
    tuple:
    - список всех палиндромов a < 100000, для которых a^2 — палиндром;
    - список всех простых p < 1000000, все циклические перестановки цифр которых
    →
    просты.
    """
    palindrom = []
    for a in range(0, 10**5):
        if is_palindrom(a) and is_palindrom(a**2):
            palindrom.append(a)

    prime = []
    for p in range(0, 10**6):
        if is_all_prime(p):
            prime.append(p)
    return (palindrom, prime)

print(palindromic_squares_and_circular_primes())
```
